## Get rid of null values in train and test

```python
train.head()
```

```python
for col in train:
    if train[col].dtype == 'object':
        train[col]= train[col].fillna("None")
for col in train:
    if train[col].dtype == 'int64' or train[col].dtype == 'float64':
        train[col]= train[col].fillna(0)
```

```python
for col in test:
    if test[col].dtype == 'object':
        test[col]= test[col].fillna("None")
for col in test:
    if test[col].dtype == 'int64' or test[col].dtype == 'float64':
        test[col]= test[col].fillna(0)
```

# Models

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
train_df0_dummy =pd.get_dummies(train_df0)
```

```python
X = train_df0_dummy.drop(columns=['SalePrice'])
y = train_df0_dummy['SalePrice']
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.3, random_state = 53)
```

```python
reg = LinearRegression().fit(X_train, y_train)

y_pred = reg.predict(X_test)

print(f'RMSE LinearRegression: {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}')
```

```python
df_pred = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_pred.head(15)
```

## XGBoost

```python
xgb = XGBRegressor(learning_rate = 0.05, n_estimators=1000, random_state=53)

xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test)

print(f'RMSE xgb: {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}')
```

```python
df_pred = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_pred.head(15)
```
